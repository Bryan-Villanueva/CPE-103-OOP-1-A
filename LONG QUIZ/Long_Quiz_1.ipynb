```python
def main():

  print("STUDENT INFORMATION", "\n")
  student_id = input("Input Student ID: ")
  name = str(input("Input Name: "))
  program = input("Input Program: ")
  course = input("Input Course: ")
  sex = str(input("Input Sex: "))
  weight = float(input("Input Weight(kg): "))
  height = float(input("Input Height(cm): "))
  print("")

  print("MIDTERM", "\n")
  quiz = float(input("Quiz: "))
  sw = float(input("Seatwork: "))
  ass = float(input("Assignment: "))
  reci = float(input("Recitation: "))

  midterm_Grade = round(quiz * 0.30 + sw * 0.30 + ass * 0.20 + reci * 0.20, 2)
  print("Midterm Grade:",midterm_Grade, "\n")

  print("FINAL GRADE", "\n")
  quiz = float(input("Quiz: "))
  sw = float(input("Seatwork: "))
  ass = float(input("Assignment: "))
  reci = float(input("Recitation: "))

  final_Grade = round(quiz * 0.30 + sw * 0.30 + ass * 0.20 + reci * 0.20, 2)
  print("Final Grade:",final_Grade, "\n")

  GWA = round((midterm_Grade + final_Grade) /2, 2)

  if GWA >= 60:
    print("GWA is:", GWA, "PASSED", "\n")

  else:
    print("GWA is:", GWA, "FAILED", "\n")

  while True:
      output = input("Type 'next' to proceed with another student or 'exit' to quit: ")
      print("")
      if output == 'next':
          main()
          break
      elif output == 'exit':
          print("Exiting program.")
          return
      else:
          print("Invalid input. Please type 'next' to proceed or 'exit' to quit.")
          print("")

main()
```
